```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1, 2, 3, 4]
y = [5, 6, 8, 5]
```

```python
plt.plot(x, y)

plt.title("Gr√°fico de linhas simples")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")

plt.show()
```

```python
plt.scatter(x, y, color="red", marker="P")

plt.title("Gr√°fico de dispers√£o simples")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")

plt.show()
```

```python
dados = np.random.randn(1000)
```

```python
plt.hist(dados, bins=30, edgecolor='black')

plt.title("Histograma simples")
plt.xlabel("Valores")
plt.ylabel("Frequ√™ncia")

plt.show()
```
