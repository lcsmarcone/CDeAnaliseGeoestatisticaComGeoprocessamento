```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Carregar o shapefile
shapefile = gpd.read_file('nordeste.shp')

# Criar o plot
fig, ax = plt.subplots(figsize=(10, 10))

# Plotar o shapefile
shapefile.plot(ax=ax, color='lightblue', edgecolor='black')

# Adicionar ttulo
ax.set_title('Mapa da Regio Nordeste do Brasil', fontsize=15)

# Mostrar o grfico
plt.show()
```

```python
# Plotar grfico de barras com as reas
fig, ax = plt.subplots(figsize=(12, 8))

# Criar o grfico de barras
shapefile['AREA_KM2'].plot(kind='bar', ax=ax, color='skyblue')

# Adicionar ttulo e rtulos
ax.set_title('Distribuio das reas dos Polgonos no Nordeste (km)', fontsize=16)
ax.set_xlabel('ndice do Polgono', fontsize=12)
ax.set_ylabel('rea (km)', fontsize=12)

# Ajustar layout para uma melhor visualizao
plt.tight_layout()

# Mostrar o grfico
plt.show()
```

```python
# Verificar se as colunas NM_UF e AREA_KM2 esto presentes
if 'NM_UF' in shapefile.columns and 'AREA_KM2' in shapefile.columns:
    # Ordenar os dados por rea
    shapefile_sorted = shapefile.sort_values(by='AREA_KM2', ascending=False)
    
    # Configurar o grfico
    fig, ax = plt.subplots(figsize=(12, 8))
    
    # Plotar o grfico de barras
    ax.bar(shapefile_sorted['NM_UF'], shapefile_sorted['AREA_KM2'], color='skyblue')
    
    # Adicionar rtulos e ttulo
    ax.set_xlabel('Unidade da Federao (UF)', fontsize=12)
    ax.set_ylabel('rea (Km)', fontsize=12)
    ax.set_title('rea por Unidade da Federao (UF) no Nordeste', fontsize=15)
    
    # Rotacionar os rtulos do eixo X para melhor visualizao
    plt.xticks(rotation=45, ha='right')
    
    # Exibir o grfico
    plt.tight_layout()
    plt.show()
else:
    print("As colunas 'NM_UF' ou 'AREA_KM2' no esto presentes no shapefile.")
```
