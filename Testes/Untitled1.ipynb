```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Carregar o shapefile
shapefile = gpd.read_file('nordeste.shp')

# Criar o plot
fig, ax = plt.subplots(figsize=(10, 10))

# Plotar o shapefile
shapefile.plot(ax=ax, color='lightblue', edgecolor='black')

# Adicionar título
ax.set_title('Mapa da Região Nordeste do Brasil', fontsize=15)

# Mostrar o gráfico
plt.show()
```

```python
# Plotar gráfico de barras com as áreas
fig, ax = plt.subplots(figsize=(12, 8))

# Criar o gráfico de barras
shapefile['AREA_KM2'].plot(kind='bar', ax=ax, color='skyblue')

# Adicionar título e rótulos
ax.set_title('Distribuição das Áreas dos Polígonos no Nordeste (km²)', fontsize=16)
ax.set_xlabel('Índice do Polígono', fontsize=12)
ax.set_ylabel('Área (km²)', fontsize=12)

# Ajustar layout para uma melhor visualização
plt.tight_layout()

# Mostrar o gráfico
plt.show()
```

```python
# Verificar se as colunas NM_UF e AREA_KM2 estão presentes
if 'NM_UF' in shapefile.columns and 'AREA_KM2' in shapefile.columns:
    # Ordenar os dados por área
    shapefile_sorted = shapefile.sort_values(by='AREA_KM2', ascending=False)
    
    # Configurar o gráfico
    fig, ax = plt.subplots(figsize=(12, 8))
    
    # Plotar o gráfico de barras
    ax.bar(shapefile_sorted['NM_UF'], shapefile_sorted['AREA_KM2'], color='skyblue')
    
    # Adicionar rótulos e título
    ax.set_xlabel('Unidade da Federação (UF)', fontsize=12)
    ax.set_ylabel('Área (Km²)', fontsize=12)
    ax.set_title('Área por Unidade da Federação (UF) no Nordeste', fontsize=15)
    
    # Rotacionar os rótulos do eixo X para melhor visualização
    plt.xticks(rotation=45, ha='right')
    
    # Exibir o gráfico
    plt.tight_layout()
    plt.show()
else:
    print("As colunas 'NM_UF' ou 'AREA_KM2' não estão presentes no shapefile.")
```
